# ***Random***

```python
random_1 = tf.random.Generator.from_seed(42) # set the seed for reproducibility
random_1 = random_1.normal(shape=(3, 2))     # create tensor from a normal distribution 
random_2 = tf.random.Generator.from_seed(42) # set the seed for reproducibility
random_2 = random_2.normal(shape=(3, 2))     # create tensor from a normal distribution 
print(random_1)
print(random_2)
random_1 == random_2    # Are they equal?
```

```python
import numpy as np
n_array = np.arange(1,13)
A = tf.constant(n_array)
B = tf.constant(n_array, shape = [2,6])
print(n_array)
print(A)
print(B)
```

* **Shape:** The length (number of elements) of each of the dimensions of a tensor.
* **Rank:** The number of tensor dimensions. A scalar has rank 0, a vector has rank 1, a matrix is rank 2, a tensor has rank n.
* **Axis** or **Dimension:** A particular dimension of a tensor.
* **Size:** The total number of items in the tensor.

```python
rank_4_tensor = tf.zeros([2, 3, 4, 5])
print(rank_4_tensor)
print("Shape, rank, size:")
rank_4_tensor.shape, rank_4_tensor.ndim, tf.size(rank_4_tensor)
```

```python
print("Datatype of every element:", rank_4_tensor.dtype)
print("Number of dimensions (rank):", rank_4_tensor.ndim)
print("Shape of tensor:", rank_4_tensor.shape)
print("Elements along axis 0 of tensor:", rank_4_tensor.shape[0])
print("Elements along last axis of tensor:", rank_4_tensor.shape[-1])
print("Total number of elements (2*3*4*5):", tf.size(rank_4_tensor).numpy()) # .numpy() converts to NumPy array
```

```python
# Get the first 2 items of each dimension
rank_4_tensor[:2, :2, :2, :3]
```

```python
# Get the last item in second matrix in first block of matrixes
# I got the last column in second matrix of first block of matrixes
print(rank_4_tensor[1, 2, :, -1])
```

```python
# Create a rank 2 tensor (2 dimensions)
# ... means use all previous dimensions
rank_2_tensor = tf.constant([[10, 7],
                             [3, 4]])
print(rank_2_tensor[..., -1])
```

There are possibilities to add more dimentions to tensors

```python
rank_3_tensor = rank_2_tensor[..., tf.newaxis]
rank_4_tensor = tf.expand_dims(rank_3_tensor, axis=-1) # "-1" means last axis
print(rank_3_tensor)
print(rank_4_tensor)
```

# **Tensor Operations**

---



```python
tensor = tf.constant([[10, 7], [3, 4]])
print(tensor + 10)

# Multiplication (known as element-wise multiplication)
print(tensor * 10)

# Use the tensorflow function equivalent of the '*' (multiply) operator (speed)
print(tf.multiply(tensor, 10))
```

```python
# Get the sqare of each value
tf.square(tensor)
```

```python
# Get the square root of each value
tf.sqrt(tf.cast(tensor, dtype=tf.float16))
```

```python
# Get the natural log of each value
tf.math.log(tf.cast(tensor, dtype=tf.float16))
```

### Matrix multiplication

```python
# Matrix multiplication in TensorFlow
# use '@' or 'tf.matmul()'
print(tensor)
print(tf.matmul(tensor, tensor))
```

**Reshape != Transpose**

```python
A = tf.constant([[1,2,3],[4,5,6]])

B = tf.transpose(A)

print(A)
print(B)
```

```python
C = tf.reshape(A, shape=[3,2])
print(C)
```

```python
# You can achieve the same result with parameters
v = tf.matmul(a=A, b=A, transpose_a=False, transpose_b=True)
print(v)
print(A@B)
```

### The dot product

```python
v1 = tf.tensordot(A, B, axes = 1)
v2 = tf.tensordot(A, B, axes = 2)
print(v1)
print(v2)
```

***Axis=0*** -- Each value of matrix A multiplies to all matrix B which results in the output matrix C, which have count of matrixes shaped as B equal to size of matrix A

***Axis=1*** -- Matrix multiplication. Row * each columns, then summs and makes first value. Then changle the row and repeat

***Axis=2*** -- Matrix A overlaps matrix B. All corresponding values are multiplied and summed to one single value

```python
A = tf.constant([1,2,3,4], shape=[2,2])
B = tf.constant([2,3,4,5], shape=[2,2])

print(A@B)
print(tf.matmul(A,B))
print(tf.tensordot(A,B, axes=1))
```

```python
print(tf.tensordot(A,B, axes=2))
print(tf.tensordot(A,B, axes=0).numpy()) # Convert to numpy array
```

**Experiments**

typically, each of these is under the *math* module, e.g. *tf.math.reduce_min()* but you can use the alias *tf.reduce_min()*

```python
A = tf.constant(np.random.randint(low=-20, high=20, size=50), dtype=tf.float32)

print(A)
print(tf.abs(A))
print(tf.reduce_max(A))
print(tf.reduce_min(A))
print(tf.reduce_mean(A))
print(tf.math.reduce_std(A))              # Can't be used withput 'math' module
print(tf.math.reduce_variance(A).numpy()) # Convert to numpy form
print(tf.reduce_sum(A).numpy())
```

```python
F = tf.constant(np.random.random(50))
print(F)
print("\n")

# Find the maximum element position of F
print(f"The maximum value of F is at position: {tf.argmax(F).numpy()}") 
print(f"The maximum value of F is: {tf.reduce_max(F).numpy()}") 
print(f"Using tf.argmax() to index F, the maximum value of F is: {F[tf.argmax(F)].numpy()}")
print(f"Are the two max values the same (they should be)? {F[tf.argmax(F)].numpy() == tf.reduce_max(F).numpy()}")
```

**Decorators** wrap a function, modifying its behavior

You can decorate your functions with ***@tf.function*** to seed it up

It turns a Python function into a callable **TensorFlow graph**

```python
@tf.function
def fun1(A_m, B_m):
  return (2*A_m + B_m)

A = tf.Variable(np.arange(1,11), dtype=tf.float16)
B = tf.Variable(np.random.rand(10), dtype=tf.float16)

print(A.numpy())
print(np.array(B))
print(fun1(A,B).numpy())
```
